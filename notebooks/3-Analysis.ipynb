## Metrics analysis

```python
metrics_ml = pd.read_csv("../data/models/results_ml_val_test.csv")
metrics_ml
```

```python
metrics_ml["overfitting"] = abs(metrics_ml["f1_val"] - metrics_ml["f1_test"])
```

```python
metrics_ml.sort_values(by=["overfitting"], ascending=[True])
```
